```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("Algerian_dataset.csv", header=1)
```

```python
df.head()
```

```python
# Data Cleaning
df.isnull().sum()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, "Region"] = 0
df.loc[122:, "Region"] = 1
```

```python
df.info()
```

```python
# Removing null values
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(122).reset_index(drop=True)
```

```python
df.columns
```

```python
# Fix spaces
df.columns = df.columns.str.strip()
```

```python
df.columns
```

```python
# Change data type to int
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

```python
# Changing other datatypes to float
objects = [features for features in df.columns if df[features].dtypes == 'O']
objects
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
# Save the cleaned Dataset
df.to_csv('Algerian_cleaned.csv', index=False)
```

```python
#drop day, month and year
df_copy = df.drop(['month', 'day', 'year'], axis=1)
```

```python
df_copy['Classes'].value_counts()
```

```python
#Encoding
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
df_copy['Classes'].value_counts()
```

```python
## Plot desnity plot for all features
plt.style.use('seaborn-v0_8')  # Works for newer matplotlib versions
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
# Percentage for Pie Chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
percentage
```

```python
#Plotting pie chart
classlabels = ["Fire", "Not Fire"]
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=classlabels,autopct='%1.1f%%')
plt.title("Pie for classes")
plt.show()
```

```python
# Fire Analysis
df['Classes'] = np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
df['Classes'].value_counts()
```

```python
# Monthly Fire Analysis
df_temp = df.loc[df['Region'] == 1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Fire Analysis of Sidi-Bel', weight = 'bold')
```

```python
# Monthly Fire Analysis
df_temp = df.loc[df['Region'] == 0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')
sns.countplot(x='month', hue='Classes', data=df)
plt.ylabel('Number of fires', weight='bold')
plt.xlabel('Months', weight='bold')
plt.title('Fire Analysis of Brjaria', weight = 'bold')
```
